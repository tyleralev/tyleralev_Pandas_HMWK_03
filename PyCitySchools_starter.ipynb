```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_schools = len(school_data_complete["school_name"].unique())

#Calculate the total number of students
total_students = school_data_complete["Student ID"].count()

#Calculate the total budget
total_budget = sum(school_data_complete["budget"].unique())

#Calculate the average math score
avg_math = school_data_complete["math_score"].mean()

#Calculate the average reading score
avg_reading = school_data_complete["reading_score"]. mean()

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
ovr_avg_score = ((avg_math  + avg_reading)/2)

#Calculate the percentage of students with a passing math score (70 or greater)
pass_math = []
pass_math = len(school_data_complete[school_data_complete["math_score"] >= 70])
pct_pass_math = (pass_math/total_students)
pct_pass_math

#Calculate the percentage of students with a passing reading score (70 or greater)
pass_reading = []
pass_reading = len(school_data_complete[school_data_complete["reading_score"] >= 70])
pct_pass_reading = (pass_reading/total_students)
pct_pass_reading

#Create a dataframe to hold the above results
# Place all of the data found into a summary DataFrame
school_summary_table = pd.DataFrame({"Total Schools": total_schools,
                              "Total Students": total_students,
                              "Total Budget": total_budget,
                              "Average Math Score": avg_math,
                              "Average Reading Score": avg_reading,
                              "% Passing Math": pct_pass_math,
                              "% Passing Reading": pct_pass_reading,
                              "Overall Passing Rate": ovr_avg_score}, index=[0])
school_summary_table
#Optional: give the displayed data cleaner formatting
```

```python
#Create an overview table that summarizes key metrics about each school, including:
cleaned_school_data = pd.DataFrame(school_data_complete, columns=['school_name','student_name','reading_score', 
                                                                 'math_score'])
cleaned_school_data = cleaned_school_data.set_index('school_name')
cleaned_school_data.head()

#grouped clean school data by school name
grouped_school_data = cleaned_school_data.groupby(['school_name'], as_index=True)
grouped_school_data


#Average Math Score
avg_math_school = grouped_school_data["math_score"].mean()
avg_math_school

#Average Reading Score
avg_reading_school = grouped_school_data["reading_score"].mean()
avg_reading_school

#% Passing Math
pct_school_math = cleaned_school_data.loc[cleaned_school_data["math_score"] >= 70]
Final_pct_school_math = pct_school_math.groupby(['school_name'], as_index=True)
Final_pct_school_math = Final_pct_school_math["math_score"].count()
Final_pct_school_math

#% Passing Reading
pct_school_reading = cleaned_school_data.loc[cleaned_school_data["reading_score"] >= 70]
Final_pct_school_reading = pct_school_reading.groupby(['school_name'], as_index=True)
Final_pct_school_reading = Final_pct_school_reading["reading_score"].count()
Final_pct_school_reading

# pcts/count Table
avgs_school_table = pd.DataFrame({"Average Math Score": avg_math_school,
                                  "Average Reading Score": avg_reading_school,
                                  "Math Pass Count": Final_pct_school_math,
                                 "Reading Pass Count": Final_pct_school_reading})
avgs_school_table.head()

#Per Student Budget
per_student_budg = school_data["budget"]/school_data["size"]
school_data["Per Student Budget"] = per_student_budg
school_data

#merged table
Final_school_data = pd.merge(school_data, avgs_school_table, how="right", on=["school_name", "school_name"])
Final_school_data.head()

#% Passing Math
pct_pass_math_school = Final_school_data["Math Pass Count"]/Final_school_data["size"]
Final_school_data["% Passing Math"] = pct_pass_math_school
Final_school_data

#% Passing Reading
pct_pass_read_school = Final_school_data["Reading Pass Count"]/Final_school_data["size"]
Final_school_data["% Passing Reading"] = pct_pass_read_school
Final_school_data

ovr_pass_rate_school = ((pct_pass_math_school+pct_pass_read_school)/2)
Final_school_data["% Overall Passing Rate"] = ovr_pass_rate_school
Final_school_data

#Create a dataframe to hold the above results
Final_schools_summary = pd.DataFrame(Final_school_data, columns=['school_name', 'type', 'size', 'budget',
       'Per Student Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math',
       '% Passing Reading', '% Overall Passing Rate'])
Final_schools_summary2 = Final_schools_summary.rename(columns = {"school_name":"School Name",
                                                                 "type":"School Type",
                                                                  "size":"Total Students",
                                                                 "budget":"Total School Budget"})

Final_schools_summary2["Total School Budget"] = Final_schools_summary2["Total School Budget"].astype(float).map("${:,.2f}".format)
Final_schools_summary2["Per Student Budget"] = Final_schools_summary2["Per Student Budget"].astype(int)
Final_schools_summary2["Average Math Score"] = Final_schools_summary2["Average Math Score"].astype(float).map("{:.2f}".format)
Final_schools_summary2["Average Reading Score"] = Final_schools_summary2["Average Reading Score"].astype(float).map("{:.2f}".format)
Final_schools_summary2["% Passing Math"] = Final_schools_summary2["% Passing Math"].astype(float).map("{:,.2%}".format)
Final_schools_summary2["% Passing Reading"] = Final_schools_summary2["% Passing Reading"].astype(float).map("{:,.2%}".format)
Final_schools_summary2["% Overall Passing Rate"] = Final_schools_summary2["% Overall Passing Rate"].astype(float).map("{:,.2%}".format)
Final_schools_summary2
```

```python
school_data_complete.head()

spend_per_student = school_data_complete

spend_per_student['Budget per student'] = spend_per_student['budget'] / spend_per_student['size']
spend_per_student.head()

spend_per_student['Spending Ranges (Per Student)'] = pd.cut(spend_per_student['Budget per student'], spending_bins, labels=group_names)
spend_per_student.head()

spend_range_group = spend_per_student.groupby('Spending Ranges (Per Student)')

total_count = spend_range_group['Budget per student'].count()
total_count.head()

spend_range_avg_math = spend_range_group['math_score'].mean()
spend_range_avg_math.head()

spend_range_avg_read = spend_range_group['reading_score'].mean()
spend_range_avg_read.head()

passing_math = spend_per_student.loc[spend_per_student['math_score'] >= 70]
passing_math_group = passing_math.groupby('Spending Ranges (Per Student)')
passing_math_count = passing_math_group['math_score'].count()

passing_reading = spend_per_student.loc[spend_per_student['reading_score'] >= 70]
passing_reading_group = passing_reading.groupby('Spending Ranges (Per Student)')
passing_reading_count = passing_reading_group['reading_score'].count()

spend_range_table = pd.DataFrame({
    'Total': total_count,
    'Average Math Score': spend_range_avg_math,
    'Average Reading Score': spend_range_avg_read,
    'Passing Math Count': passing_math_count,
    'Passing Reading Count': passing_reading_count
})

spend_range_table.head()

spend_range_table['% Passing Math'] = (spend_range_table['Passing Math Count'] / spend_range_table['Total']) * 100

spend_range_table['% Passing Reading'] = (spend_range_table['Passing Reading Count'] / spend_range_table['Total']) * 100

spend_range_table['% Overall Passing Rate'] = (spend_range_table['% Passing Math'] + spend_range_table['% Passing Reading']) / 2

spend_df = spend_range_table[['Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing Rate']]
spend_df
```
